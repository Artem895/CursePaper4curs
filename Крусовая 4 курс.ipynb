```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
##import math
```

```python
## создание случайны гамма для каждого ансамбля 
final_gamma_df=pd.DataFrame()
for i in range(0,9):
    array_positive = np.random.uniform(0, 1, 7)
    array_antypositive=np.random.uniform(-1, 0, 2)
    gammadf=pd.DataFrame(array_positive)
    gammadf1=pd.DataFrame(array_antypositive)
    gammadf=gammadf.append(gammadf1,ignore_index=True)
    final_gamma_df=final_gamma_df.append(gammadf,ignore_index=True)
```

```python
def func(x,y):
    return x+(x**(1/3))-y
```

```python
## метода половинного деления 
x1=0
x2=50
eps=0.1
c=25
def method_poldel(y,x1,x2,c):
    eps=0.1
    
    if (func(x1,y)*func(x2,y))<0:
        while func(c,y)>eps:
            c=(x2-x1)/2
            if func(x1,y)*func(c,y)<0:
                x2=c
            elif func(c,y)*func(x2,y)<0:
                x1=c
        
    else:
        print(x1,"x1 ",x2,"x2 ",c,"c ",y,"y")
        x2+=10000
        print(x2)
        method_poldel(y,x1,x2,c)
        
    return (x2-x1)/2        
        
```
