# Необходимые функции

```python
def get_cube_root(x):
    if x < 0:
        x = abs(x)
        cube_root = x**(1/3)*(-1)
    else:
        cube_root = x**(1/3)
    return cube_root
```

```python
def get_c(x1,x2):
    c=(x1+x2)/2.
    return c   
```

```python
def get_from_array(n,m):
    l=9
    global arrray_of_x_old
    if (n<0)|(m<0)|(n>=l)|(m>=l):
        return 0
    else:
        return array_of_x_old[n][m]
```

```python
def method_poldel_2(y,x1,x2,n,m,k):
    eps=0.001
    if (func_with_connection(x1,y,n,m,k)*func_with_connection(x2,y,n,m,k))<0:
        c=get_c(x1,x2)
        while abs(func_with_connection(c,y,n,m,k))>eps:
            c=get_c(x1,x2)
            if func_with_connection(x1,y,n,m,k)*func_with_connection(c,y,n,m,k)<0:
                x2=c
            elif func_with_connection(c,y,n,m,k)*func_with_connection(x2,y,n,m,k)<0:
                x1=c
        
    else:
        x1-=10.
        x2+=10.
        method_poldel_2(y,x1,x2,n,m,k)       
    return get_c(x1,x2)     
```

# Наша функция


```python
def func_with_connection(x,y,n,m,k):
    return (x+(get_cube_root(x))-y)+k*get_cube_root(get_from_array(n-1,m))+k*get_cube_root(get_from_array(n+1,m))\
            +k*get_cube_root(get_from_array(n,m-1))+k*get_cube_root(get_from_array(n,m+1))
```
